# Пропущенные значения

```python
from random import randint

def add_nan(row):
    rand = randint(0, row.size-1)
    row.iloc[rand] = None
    return row

nan_df = df.sample(20).apply(add_nan, axis=1)

nan_df.shape
```

```python
nan_df
```

```python
nan_df.isna().sum()
```

```python
nan_df.dropna()
```

```python
nan_df.dropna(axis=1)
```

```python
nan_df.fillna(0) #nan_df.fillna('NA'), nan_df.fillna(-1)
```

```python
nan_df.fillna(method="bfill")
```

```python
numeric_columns = nan_df.select_dtypes(include='number').columns.tolist()
object_columns = nan_df.select_dtypes(exclude='number').columns.tolist()


mean_values = nan_df[numeric_columns].mean()
mean_values
```

```python
nan_df = nan_df.fillna(mean_values)
nan_df
```

```python
nan_df.isna().sum()
```

```python
nan_df[object_columns].describe()
```

```python
nan_df[object_columns].describe().loc['top']
```

```python
most_frequent_objects = nan_df[object_columns].describe().loc['top']
nan_df = nan_df.fillna(most_frequent_objects)
nan_df
```
